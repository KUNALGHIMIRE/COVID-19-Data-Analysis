```python

# List of countries
countries = ['Nepal', 'India', 'US']

for country in countries:
    # Filter data for the country
    country_df = df[df['Country/Region'] == country]
    
    # Group by Date and sum confirmed and deaths
    daily = country_df.groupby('Date')[['Confirmed', 'Deaths']].sum()
    
    # Plot cumulative confirmed vs deaths
    plt.figure(figsize=(10,4))
    plt.plot(daily.index, daily['Confirmed'], label='Confirmed Case', color='blue')
    plt.plot(daily.index, daily['Deaths'], label='Deaths', color='red')
    plt.title(f'COVID-19 Confirmed vs Deaths in {country}')
    plt.xlabel('Date')
    plt.ylabel('Number of Cases')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Calculate correlation
    correlation = daily['Confirmed'].corr(daily['Deaths'])
    print(f'{country} correlation between Confirmed and Deaths: {correlation:.2f}')
```

```python
countries = ['India', 'Nepal', 'US']

for c in countries:
    # Filter country data
    country = df[df['Country/Region'] == c]
    
    # Sum Confirmed and Deaths for the country
    total_confirmed = country['Confirmed'].sum()
    total_deaths = country['Deaths'].sum()
    
    # Calculate Case Fatality Rate
    fatality_rate = (total_deaths / total_confirmed) * 100
    
    print(f"{c} CFR: {fatality_rate:.2f}%")
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Make sure your Date column is datetime
df3['Date'] = pd.to_datetime(df3['Date'])

# -------------------------------
# 1️⃣ Find Top 5 countries by total confirmed cases
# -------------------------------
top5_countries = df3.groupby('Country/Region')['Confirmed'].sum() \
                    .sort_values(ascending=False) \
                    .head(5) \
                    .index.tolist()  # <-- add parentheses here

print("Top 5 countries:", top5_countries)

# -------------------------------
# 2️⃣ Plot daily new confirmed cases
# -------------------------------
plt.figure(figsize=(14,7))

for country in top5_countries:
    # Filter data for the country
    country_data = df3[df3['Country/Region'] == country].groupby('Date')['Confirmed'].sum()
    
    # Calculate daily new cases
    daily_new = country_data.diff().fillna(0)
    
    # Plot the line
    plt.plot(daily_new.index, daily_new, label=country)

# -------------------------------
# 3️⃣ Customize plot
# -------------------------------
plt.title('Daily New Confirmed COVID-19 Cases (Top 5 Countries)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Daily New Cases', fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```
